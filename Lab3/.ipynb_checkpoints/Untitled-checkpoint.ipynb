```python
all_df = pd.read_csv("column_2C.csv",header=0,names=['IOP', 'PI', 'LLA', 'SS','PR','DOLS','Category'],index_col=False)
all_df.head()
```

```python
def load_data():
    column_names=['IOP', 'PI', 'LLA', 'SS','PR','DOLS','Category']
    column_2C = pd.read_csv("column_2C.csv",names=column_names)
    values = column_2C.drop("Category", axis=1)
    data = pd.DataFrame(column_2C, columns=['Category'])
    data.loc[data['Category'] == 'AB'] = 1
    data.loc[data['Category'] == 'NO'] = 0
    result = pd.concat([values, data], axis=1)
    data_train= result.values.astype(np.float32)
    np.random.shuffle(data_train)    
   
    return data_train
```

```python
data = load_data() 
print(data.iloc[:,-1])
sep = int(0.7*len(data))
print(sep)
train_data = data[:sep]
test_data = data[sep:]
print(train_data)
print("111")
print(test_data)
```

```python
# assign features to X
X = all_df.drop("Category", axis=1)

from sklearn.decomposition import PCA

#feature_names = list(X.columns)
Xs_pca =PCA(2).fit_transform(X)
```

```python
PCA_df = pd.DataFrame()
PCA_df['PCA_1'] = Xs_pca[:,0]
PCA_df['PCA_2'] = Xs_pca[:,1]

# visualise the Malignant and benign using the two PCA features

plt.figure(figsize = (6,6))
plt.plot(PCA_df['PCA_1'][all_df['Category'] == 'AB'],PCA_df['PCA_2'][all_df['Category'] == 'AB'],'ro',alpha=0.7,)
plt.plot(PCA_df['PCA_1'][all_df['Category'] == 'NO'],PCA_df['PCA_2'][all_df['Category'] == 'NO'],'bo',alpha=0.7)

plt.xlabel('PCA_1')
plt.ylabel('PCA_2')
plt.legend(['Malignant','Benign'])
```
